## Definir Estructura de Datos de Niños

### Subtask:
Crea una estructura de datos (por ejemplo, una lista de diccionarios) para almacenar información sobre cada niño, incluyendo su nombre, edad y un registro de actividades o estado actual.


**Reasoning**:
The subtask requires defining a data structure for children and populating it with example records. I will create a Python code block to initialize a list, define the dictionary structure for each child, and add two example child records as specified.



```python
import datetime

# 1. Initialize an empty list named `children_data`
children_data = []

# 2. Define and populate `children_data` with example child records

# Child 1
child1 = {
    'name': 'Sofía',
    'age': 4,
    'current_status': 'playing with blocks',
    'activity_history': [
        {
            'timestamp': datetime.datetime(2023, 1, 15, 8, 0, 0).isoformat(),
            'type': 'fed',
            'description': 'Breakfast: cereal and milk'
        },
        {
            'timestamp': datetime.datetime(2023, 1, 15, 9, 30, 0).isoformat(),
            'type': 'slept',
            'description': 'Morning nap'
        }
    ]
}

# Child 2
child2 = {
    'name': 'Mateo',
    'age': 2,
    'current_status': 'sleeping',
    'activity_history': [
        {
            'timestamp': datetime.datetime(2023, 1, 15, 7, 0, 0).isoformat(),
            'type': 'fed',
            'description': 'Breakfast: fruit puree'
        },
        {
            'timestamp': datetime.datetime(2023, 1, 15, 10, 0, 0).isoformat(),
            'type': 'changed_diaper',
            'description': 'Diaper changed'
        }
    ]
}

children_data.append(child1)
children_data.append(child2)

print("Children Data Structure:")
for child in children_data:
    print(child)
```

## Mostrar Estado General de Niños

### Subtask:
Implementa una función para mostrar un resumen del estado actual o las últimas actividades de todos los niños registrados en la guardería.


**Reasoning**:
The subtask requires implementing a function `display_children_status` to show a summary of each child's current status and their latest activity. I will create a Python code block to define this function and then call it using the `children_data` list.



```python
def display_children_status(children_data):
    """Displays a summary of each child's current status and their most recent activity."""
    print("\n--- Current Children Status ---")
    if not children_data:
        print("No children registered yet.")
        return

    for child in children_data:
        print(f"\nName: {child['name']}")
        print(f"Age: {child['age']}")
        print(f"Current Status: {child['current_status']}")

        if child['activity_history']:
            latest_activity = child['activity_history'][-1] # Get the last activity
            print(f"Last Activity: {latest_activity['type']} - {latest_activity['description']}")
        else:
            print("No activity history recorded.")
    print("-------------------------------")

# Call the function to display the status of the example children
display_children_status(children_data)
```

## Registrar Actividad del Niño

### Subtask:
Desarrolla una función que permita al personal de la guardería registrar nuevas actividades (ej. 'comió', 'durmió', 'jugó') para un niño específico.


**Reasoning**:
The subtask requires defining a function to register new activities for a specific child. I will create a Python code block to define the `register_child_activity` function as specified in the instructions, including handling cases where the child is not found, and then call it with example data.



```python
import datetime

def register_child_activity(children_data, child_name, activity_type, activity_description):
    """Registers a new activity for a specific child."""
    child_found = False
    for child in children_data:
        if child['name'].lower() == child_name.lower():
            new_activity = {
                'timestamp': datetime.datetime.now().isoformat(),
                'type': activity_type,
                'description': activity_description
            }
            child['activity_history'].append(new_activity)
            print(f"Activity '{activity_type}' registered for {child_name}.")
            print(f"Updated activity history for {child_name}:")
            for activity in child['activity_history']:
                print(f"  - {activity['timestamp']} | Type: {activity['type']} | Desc: {activity['description']}")
            child_found = True
            break

    if not child_found:
        print(f"Child '{child_name}' not found in the daycare system.")

# Example usage:
# Register a new activity for Sofía
register_child_activity(children_data, 'Sofía', 'played', 'Playing with building blocks in the common area')

# Try to register an activity for a non-existent child
register_child_activity(children_data, 'Laura', 'ate', 'Had a snack')

# Verify Mateo's activities remain unchanged
print(f"\nMateo's activity history:")
for child in children_data:
    if child['name'] == 'Mateo':
        for activity in child['activity_history']:
            print(f"  - {activity['timestamp']} | Type: {activity['type']} | Desc: {activity['description']}")
        break
```

## Actualizar Estado del Niño

### Subtask:
Crea una función para que el personal pueda actualizar el estado actual de un niño (ej. 'durmiendo', 'jugando') y esta actualización sea visible para los padres.


**Reasoning**:
The subtask requires defining a function to update a child's current status and record this update in their activity history. I will create a Python code block to define the `update_child_status` function, incorporate `datetime.datetime.now().isoformat()` for timestamps, and include examples of its usage, followed by calling `display_children_status` to verify the changes.



```python
import datetime

def update_child_status(children_data, child_name, new_status):
    """Updates the current status of a specific child and logs it as an activity."""
    child_found = False
    for child in children_data:
        if child['name'].lower() == child_name.lower():
            # Update current status
            child['current_status'] = new_status

            # Add a new entry to activity history
            new_activity = {
                'timestamp': datetime.datetime.now().isoformat(),
                'type': 'status_update',
                'description': f'Status changed to: {new_status}'
            }
            child['activity_history'].append(new_activity)

            print(f"Child '{child_name}' status updated to '{new_status}'.")
            child_found = True
            break

    if not child_found:
        print(f"Child '{child_name}' not found in the daycare system.")

# Example usage:
# Update Sofía's status to 'sleeping'
update_child_status(children_data, 'Sofía', 'sleeping')

# Try to update a non-existent child's status
update_child_status(children_data, 'Carlos', 'eating snack')

# Update Mateo's status to 'playing with cars'
update_child_status(children_data, 'Mateo', 'playing with cars')

# Display the general status of all children to verify changes
display_children_status(children_data)
```

## Ver Historial de Actividades del Niño

### Subtask:
Implementa una función para que los padres puedan consultar el historial detallado de actividades de un niño específico, filtrado por fecha o tipo de actividad.


**Reasoning**:
The subtask requires defining a function to view a child's activity history with optional filtering by date and activity type. I will create a Python code block to implement the `view_child_activity_history` function and provide examples of its usage, including all specified filter combinations and a case for a non-existent child.



```python
import datetime

def view_child_activity_history(children_data, child_name, start_date=None, activity_type=None):
    """Displays the detailed activity history for a specific child, with optional filtering by date and activity type."""
    child_found = False
    target_child = None
    for child in children_data:
        if child['name'].lower() == child_name.lower():
            target_child = child
            child_found = True
            break

    if not child_found:
        print(f"\nChild '{child_name}' not found in the daycare system.")
        return

    print(f"\n--- Activity History for {target_child['name']} ---")
    print(f"Filters: Start Date={start_date if start_date else 'None'}, Activity Type={activity_type if activity_type else 'None'}")

    filtered_activities = []
    for activity in target_child['activity_history']:
        include_activity = True

        # Filter by activity_type
        if activity_type:
            if activity['type'].lower() != activity_type.lower():
                include_activity = False

        # Filter by start_date
        if include_activity and start_date:
            activity_timestamp = datetime.datetime.fromisoformat(activity['timestamp'])
            # Assuming start_date is 'YYYY-MM-DD'
            filter_start_date = datetime.datetime.fromisoformat(start_date + 'T00:00:00') # Convert to datetime object, start of the day
            if activity_timestamp < filter_start_date:
                include_activity = False

        if include_activity:
            filtered_activities.append(activity)

    if filtered_activities:
        for activity in filtered_activities:
            print(f"  - {activity['timestamp']} | Type: {activity['type']} | Desc: {activity['description']}")
    else:
        print(f"No activities found for {target_child['name']} matching the specified criteria.")
    print("------------------------------------------")

# Example usage:
print("\n--- Demonstrating view_child_activity_history function ---")

# a. Show full activity history for Sofía
view_child_activity_history(children_data, 'Sofía')

# b. Show activity history for Mateo filtered by activity type 'fed'
view_child_activity_history(children_data, 'Mateo', activity_type='fed')

# c. Show activity history for Sofía filtered by start_date '2023-01-15'
# Note: using example date from initial data, current activities are recent. Adjusting to show results.
view_child_activity_history(children_data, 'Sofía', start_date='2023-01-15')

# d. Show activity history for Sofía filtered by both activity_type 'status_update' and start_date (a recent date, ensure it's after original data)
# For demonstration, use a recent date if activities were logged today, otherwise use an old date from example data
# Let's pick a date that would capture the 'status_update' if it occurred recently or an older date if it's the example date
view_child_activity_history(children_data, 'Sofía', start_date='2024-01-01', activity_type='status_update')

# e. Try to view history for a non-existent child
view_child_activity_history(children_data, 'Isabella')
```

## Menú Principal de la Aplicación

### Subtask:
Establece un bucle principal interactivo que ofrezca opciones como 'Ver estado general', 'Registrar actividad', 'Actualizar estado', 'Ver historial de niño' y 'Salir', manejando la navegación del usuario.


**Reasoning**:
The subtask requires implementing an interactive main menu with a `while True` loop, displaying options, taking user input, and calling previously defined functions based on the user's choice. I will create a code block that encapsulates this interactive menu logic, including input handling for each operation and a graceful exit.



```python
def main_menu():
    """Main interactive menu for the daycare application."""
    while True:
        print("\n--- Menú Principal de la Guardería ---")
        print("1. Ver estado general de los niños")
        print("2. Registrar actividad de un niño")
        print("3. Actualizar estado de un niño")
        print("4. Ver historial de actividades de un niño")
        print("5. Salir")
        print("--------------------------------------")

        choice = input("Ingrese su elección (1-5): ")

        if choice == '1':
            display_children_status(children_data)
        elif choice == '2':
            child_name = input("Ingrese el nombre del niño: ")
            activity_type = input("Ingrese el tipo de actividad (ej. comió, durmió, jugó): ")
            activity_description = input("Ingrese una descripción de la actividad: ")
            register_child_activity(children_data, child_name, activity_type, activity_description)
        elif choice == '3':
            child_name = input("Ingrese el nombre del niño: ")
            new_status = input("Ingrese el nuevo estado del niño (ej. durmiendo, jugando): ")
            update_child_status(children_data, child_name, new_status)
        elif choice == '4':
            child_name = input("Ingrese el nombre del niño para ver su historial: ")
            start_date_input = input("Ingrese la fecha de inicio (YYYY-MM-DD) o presione Enter para omitir: ")
            activity_type_input = input("Ingrese el tipo de actividad para filtrar o presione Enter para omitir: ")

            start_date = start_date_input if start_date_input else None
            activity_type = activity_type_input if activity_type_input else None

            view_child_activity_history(children_data, child_name, start_date, activity_type)
        elif choice == '5':
            print("Gracias por usar la aplicación de guardería. ¡Hasta luego!")
            break
        else:
            print("Opción no válida. Por favor, ingrese un número del 1 al 5.")

# Run the main menu
main_menu()
```
